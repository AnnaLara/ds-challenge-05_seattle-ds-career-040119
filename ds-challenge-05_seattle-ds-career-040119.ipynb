```python
import pandas as pd

attendance_df = pd.read_csv('data.csv', names=('Apr1', 'Apr2', 'Apr3', 'Apr4', 'Apr5', 'Apr8'))
attendance_df.head(30)
```

```python
hour_df = attendance_df.applymap(lambda x: x[0])
```

```python
hour_df = hour_df.replace(to_replace = 'a', value = 0)
```

```python
hour_df = hour_df.applymap(lambda x: int(x))
```

```python
hour_df.info()
```

```python
values = hour_df.values
```

```python
all_values = []
for li in values:
    for v in li:
        all_values.append(v)
        
all_values
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.hist(all_values, bins=30)
plt.xlabel('Hour of arrival');
plt.ylabel('Frequency');
plt.xlim(5, 10);
plt.title('Hour of arrival to Flatiron')
plt.show()
```

```python
##from Miles's solution:




def convert_time_to_minutes(time_text: str) -> int:
    """Converts time formatted as 'H:MM AM' to minutes before/after 9:00 AM"""
    try:
        time_delta = pd.Timestamp(time_text) - pd.Timestamp("9:00 am")
        return int(time_delta.total_seconds() / 60)
    except ValueError:
        return None
```

```python
data = attendance_df.values.reshape(-1)
data
```

```python
minutes_early_or_late = [convert_time_to_minutes(time_text) for time_text in data]
```

```python
minutes_early_or_late
```

```python
y = [value for value in minutes_early_or_late if value is not None]  # drop missing/None values
```

```python
x_label = 'Minutes (negative is early, zero is on time, positive is late)'
y_label = 'Count'
hist_title = 'Student On-Time Performance'
```

```python
fig, ax = plt.subplots()
ax.hist(y)  # Use the default number of bins
ax.set_xlabel(x_label)
ax.set_ylabel(y_label)
fig.suptitle(hist_title)
plt.show()
```

```python

fig, ax = plt.subplots()
ax.hist(y, bins=50)  # More bins is better, right?
ax.set_xlabel(x_label)
ax.set_ylabel(y_label)
fig.suptitle(hist_title)
plt.show()
```
