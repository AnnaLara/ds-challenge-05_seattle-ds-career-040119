```python
import pandas as pd

attendance_df = pd.read_csv('data.csv', names=('Apr1', 'Apr2', 'Apr3', 'Apr4', 'Apr5', 'Apr8'))
attendance_df.head(30)
```

```python
attendance_df.info()
```

```python
attendance_df = attendance_df.replace(to_replace = 'absent', value = '0')
attendance_df
```

```python
hour_df = attendance_df.applymap(lambda x: x[0])
```

```python
hour_df
```

```python
hour_df.applymap(lambda x: int(x))
```

```python
apr1_attendance = hour_df['Apr1'].value_counts()
```

```python
apr1_attendance.T
```

```python
ax = apr1_attendance.plot.hist()
ax.set_xlabel('x_label')
ax.set_ylabel('y_label')
```
