```python
import pandas as pd

attendance_df = pd.read_csv('data.csv', names=('Apr1', 'Apr2', 'Apr3', 'Apr4', 'Apr5', 'Apr8'))
attendance_df.head(30)
```

```python
attendance_df.info()
```

```python
attendance_df = attendance_df.replace(to_replace = 'absent', value = '0')
attendance_df
```

```python
attendance_df.describe()
```

```python
hour_df = attendance_df.applymap(lambda x: x[0])
```

```python
hour_df
```

```python
hour_df = hour_df.replace(to_replace = 'a', value = 0)
```

```python
hour_df = hour_df.applymap(lambda x: int(x))
```

```python
hour_df.info()
```

```python
values = hour_df.values
```

```python
all_values = []
for li in values:
    for v in li:
        all_values.append(v)
        
all_values
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.hist(all_values, bins=5)
plt.xlabel('Hour of arrival');
plt.ylabel('Frequency');
plt.xlim(0, 15);
plt.title('Hour of arrival to Flatiron')
plt.show()
```
